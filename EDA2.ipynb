```python
# Data Exploration and Preprocessing

import pandas as pd
data = pd.read_csv('/content/adult_with_headers.csv')
data
```

```python
data.info()
```

```python
data.describe(include='all')
```

```python
#missing_values
data.isnull().sum()
```

```python
for column in data.columns:
  if data[column].isnull().sum()>0:
    if data[column].dtype=='object':
      data[column].fillna(data[column].mode()[0],inplace=True)
    else:
      data[column].fillna(data[column].mean(),inplace=True)
```

```python
#	Apply scaling techniques to numerical features:
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler_standard = StandardScaler()
data_standard_scaled = data.copy()
data_standard_scaled[data_standard_scaled.select_dtypes(include=['int64', 'float64']).columns] = scaler_standard.fit_transform(data_standard_scaled.select_dtypes(include=['int64', 'float64']))
data_standard_scaled
```

```python
scaler_minmax=MinMaxScaler()
data_minmax_scaled=data.copy()
data_minmax_scaled[data_minmax_scaled.select_dtypes(include=['int64','float64']).columns]=scaler_minmax.fit_transform(data_minmax_scaled.select_dtypes(include=['int64','float64']))
data_minmax_scaled
```

```python
# 2.Encoding Techniques
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
categorical_columns = [col for col in data.select_dtypes(include=['object']).columns if data[col].nunique() < 5]
data_one_hot_encoded = pd.get_dummies(data, columns=categorical_columns)
data_one_hot_encoded
```

```python
label_encoder = LabelEncoder()
for col in data.select_dtypes(include=['object']).columns:
    if col not in categorical_columns:
        data[col] = label_encoder.fit_transform(data[col])
label_encoder
```

```python
# Example of creating new features
data['hours_per_week_squared'] = data['hours_per_week'] ** 2
data['age_income_ratio'] = data['age'] / (data['capital_gain'] + 1)
data['hours_per_week_squared'],data['age_income_ratio']
```

```python
#Apply a transformation
import numpy as np
data['capital_gain'] = np.log1p(data['capital_gain'])
data['capital_gain']
```

```python
# Feature Selection

from sklearn.ensemble import IsolationForest
iso_forest = IsolationForest(contamination=0.1)
outliers = iso_forest.fit_predict(data.select_dtypes(include=['int64', 'float64']))
data_no_outliers = data[outliers == 1]
iso_forest
```
